```python
import vaex
import pandas as pd
df_vaex = vaex.open('neighbors.hdf5')
df_pandas = df_vaex.to_pandas_df()
df_neighbors = df_pandas.sort_values(by='ebv_bstep', ascending=True)
```

```python
import numpy as np
start = np.linspace(0, int(8000), 5, endpoint = True)
finish = np.linspace(int(2000), int(10000), 5, endpoint = True)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
df_testing = df_pandas 
for j in tqdm(range(len(start))):
  column_list = df_pandas[int(start[j]):int(finish[j])] #column_list = number of object
  result = []
  for i in tqdm(range(len(column_list))):
    # "Some Code" that returns a modified dataframe of each File
    #modified_df = pd.DataFrame({'values': np.random.randint(1, 5, size=np.random.randint(10))})
    df_t = df_neighbors[df_neighbors['teff'].between(df_testing.teff[i]-200, df_testing.teff[i]+200)]
    df_t = df_t[df_t['logg'].between(df_testing.logg[i]-0.18, df_testing.logg[i]+0.18)]
    df_t = df_t[df_t['fe_h'].between(df_testing.fe_h[i]-0.16, df_testing.fe_h[i]+0.16)]
    df_t = df_t[df_t['ebv_bstep'] < 0.1]
    df_t = df_t.sort_values(by = 'sobject_id_1', na_position='last')
    df_t = df_t['sobject_id_1']
    #print(df_t)
    #print(str(i))
    #print('---')
    #append the `values` column to a list
    result.append(pd.Series(df_t, name=str(i)))
  result = pd.concat(result, axis=1)
  result.to_hdf('neighbors_' + str(j) + '.h5', key = 'df')
```
